```python
import psycopg2, random
from faker import Faker
from datetime import timedelta, date

fake = Faker()
conn = psycopg2.connect(
    dbname="Ecommerce", user="postgres",
    password="SurajSitoula@1", host="localhost"
)
cur = conn.cursor()

# 200 customers
for _ in range(200):
    cur.execute(
        "INSERT INTO customers (name,email,city,signup_date)"
        " VALUES (%s,%s,%s,%s)",
        (fake.name(), fake.unique.email(),
         fake.city(), fake.date_between('-2y','today'))
    )

# 50 products
cats = ['Electronics','Clothing','Home','Beauty','Sports']
for _ in range(50):
    cur.execute(
        "INSERT INTO products (product_name,category,unit_price)"
        " VALUES (%s,%s,%s)",
        (fake.word().capitalize()+' '+fake.word().capitalize(),
         random.choice(cats), round(random.uniform(10,500),2))
    )

# 10,000 orders
base = date(2023,1,1)
for _ in range(10000):
    cid = random.randint(1,200)
    pid = random.randint(1,50)
    qty = random.randint(1,5)
    cur.execute(
        "SELECT unit_price FROM products WHERE product_id=%s",(pid,))
    price = cur.fetchone()[0]
    odate = base + timedelta(days=random.randint(0,730))
    cur.execute(
        "INSERT INTO orders"
        " (customer_id,product_id,quantity,order_date,total_amount)"
        " VALUES (%s,%s,%s,%s,%s)",
        (cid, pid, qty, odate, round(price*qty,2))
    )

conn.commit()
cur.close()
conn.close()
print("Done — 10,000 orders inserted")
```
